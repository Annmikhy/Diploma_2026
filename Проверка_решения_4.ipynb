```python
#проверка решения 4 с параметрами из ALGEBRAIC SOLUTIONS OF THE SIXTH PAINLEV E EQUATION OLEG LISOVYY AND YURIY TYKHYY
from sympy import *

s = symbols('s')

# Параметры
theta_x = Rational(1, 2)
theta_y = Rational(1, 4)
theta_z = Rational(1, 2)
theta_inf = Rational(2, 3)

# w и t
w = (Rational(9, 1) * s * (2*s**3 - 3*s + 4)) / (4 * (s+1) * (s-1)**2 * (2*s**2 + 6*s + 1))
t = 27 * s**2 / (4 * (s**2 - 1)**3)

print("Проверка Solution 4 уравнения Пенлеве VI")
print(f"θ_x = {theta_x}, θ_y = {theta_y}, θ_z = {theta_z}, θ_∞ = {theta_inf}")

# производные dw/dt и d²w/dt² через параметр s
dw_ds = diff(w, s)
dt_ds = diff(t, s)

# dw/dt = (dw/ds) / (dt/ds)
dw_dt = simplify(dw_ds / dt_ds)

# d²w/dt² = d/dt(dw/dt) = [d/ds(dw/dt)] / (dt/ds)
d2w_dt2 = simplify(diff(dw_dt, s) / dt_ds)

# ---------------------------------------------------------
# Левая часть уравнения -- L = w''
# ---------------------------------------------------------
lhs = d2w_dt2
# ---------------------------------------------------------
# Правая часть уравнения -- R
# ---------------------------------------------------------
A = Rational(1,2) * (1/w + 1/(w-1) + 1/(w-t)) * dw_dt**2

B = (1/t + 1/(t-1) + 1/(w-t)) * dw_dt

C = w*(w-1)*(w-t) / (2*t**2 * (t-1)**2) * (
    (theta_inf-1)**2
    - theta_x**2 * t / w**2
    + theta_y**2 * (t-1) / ((w-1)**2)
    + (1 - theta_z**2) * t*(t-1) / ((w-t)**2)
)

rhs = A - B + C
# ---------------------------------------------------------
# Упрощение левой и правой частей
# ---------------------------------------------------------
lhs_simplified = simplify(lhs)
rhs_simplified = simplify(rhs)
print()

# ---------------------------------------------------------
# Вывод
# ---------------------------------------------------------
print("Результат:")
print("\n LHS(s) (упрощённая левая часть):")
print(lhs_simplified)
print("\n RHS(s) (упрощённая правая часть):")
print(rhs_simplified)
print()

# Разность
residual = simplify(lhs_simplified - rhs_simplified)

if residual == 0:
    print("Разность равна нулю")
    print(" Solution 4 является верным решением PVI")
else:
    print("Разность не равна нулю")
    print(" неверное решение")
    print("\nРазность =", residual)

# ---------------------------------------------------------
# подстановка конкретного s
# ---------------------------------------------------------

print("\nДополнительная проверка: подставим числовое значение s = 2")

s_val = Rational(2, 1)

lhs_num = lhs_simplified.subs(s, s_val)
rhs_num = rhs_simplified.subs(s, s_val)
res_num = residual.subs(s, s_val) if residual != 0 else 0

print(f"При s = {s_val}:")
print(f"  LHS = {lhs_num}")

# Проверяем, является ли число комплексным
if lhs_num.is_real:
    print(f"  LHS ≈ {float(lhs_num):.6f}")
else:
    # Для комплексных чисел используем evalf()
    lhs_complex = complex(lhs_num.evalf())
    print(f"  LHS ≈ {lhs_complex:.6f}")
    print(f"  (вещественная часть: {lhs_complex.real:.6f}, мнимая: {lhs_complex.imag:.6f})")

print(f"  RHS = {rhs_num}")
if rhs_num.is_real:
    print(f"  RHS ≈ {float(rhs_num):.6f}")
else:
    rhs_complex = complex(rhs_num.evalf())
    print(f"  RHS ≈ {rhs_complex:.6f}")
    print(f"  (вещественная часть: {rhs_complex.real:.6f}, мнимая: {rhs_complex.imag:.6f})")

print(f"  Разность (LHS - RHS) = {res_num}")
if res_num != 0 and not res_num.is_real:
    res_complex = complex(res_num.evalf())
    print(f"  Разность ≈ {res_complex:.6f}")
```
